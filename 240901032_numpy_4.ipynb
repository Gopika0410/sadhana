```python
import numpy as np
arr1=[10,20,30,40]
arr2=[2,4,5,8,10]
a=np.array(arr1)
b=np.array(arr2)
print("\nOriginal Array")
print(a)
print(b)
print("\nVector addition")
print(a+b)
print("\nVector subtraction")
print(a-b)
print("\nVector multiplication")
print(a*b)
print("\nVector division")
print(a/b)
print(a.dot(b))
print("\nscalar multiplication")
sclr=5
print("scalar value=",sclr)
print("array=",a)
print("result=",a*sclr)
```
